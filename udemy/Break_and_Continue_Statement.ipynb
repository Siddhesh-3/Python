```
# Break

list = [1,2,"stop",2.5]

for i in list:
  print(i)
  if i == "stop":
    print("Loop stop here")
    break
```

```
my_list = [ 1 , 2 , 3 , "stop" , 5 , 4.1]

for i in my_list:
  print(i)
  if i == "stop":
    print("stop here")
    break
print("this is outside the loop")    
```

```
my_list = [ 1 , 2 , 3 , "stop" , 5 , 4.1]
index = 0
while index < len(my_list):
  listitem = my_list[index]
  print(listitem)
  if listitem == "stop":
      print("The loop will stop !")
      break
  index = index + 1
```

```
counter = 3
while counter <= 6:
    print(counter)
    if counter == 1:
        break
    counter -=1
```

```
list_ages = [10 , 11 , 19 , 7 , 8 , 24 , 91 , 74 , 25 , 53 , 41 , 50 , 63 , 86]
ages_even = []
for current_age in list_ages:
    if current_age % 2 == 0:
        ages_even.append(current_age) #The append() method appends an element to the end of the list.
        print(ages_even)
        continue
```
